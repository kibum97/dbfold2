```python
import os
import json
import pandas as pd
from collections import defaultdict
```

```python
will_to_edo_map = {
    1: 7,  2: 0,  3: 19, 4: 10, 5: 9,
    6: 15, 7: 16, 8: 3,  9: 6,  10: 2,
    11: 5, 12: 11, 13: 1, 14: 4, 15: 12,
    16: 8, 17: 13, 18: 18, 19: 17, 20: 14
}

def get_will2edo(edo):
    # .get(key, default) handles the "else return 0" logic
    return will_to_edo_map.get(edo, 0)
```

```python
for ncluster in [20, 30]:
    cluster_file = f'/n/home01/kibumpark/pkg/dbfold2/MCPU/src_cpp/config_files/center{ncluster}_state'
    data = defaultdict(list)
    with open(cluster_file, 'r') as f:
        lines = f.readlines()
    for i, line in enumerate(lines):
        c_label, x, y = line.split()
        c_label = get_will2edo(int(c_label))
        x = float(x)
        y = float(y)
        data['cluster_index'].append(i%ncluster)
        data['residue_index'].append(c_label)
        data['x'].append(x)
        data['y'].append(y)
    df = pd.DataFrame(data)
    output_file = f'/n/home01/kibumpark/pkg/dbfold2/MCPU/src_cpp/config_files/center{ncluster}_state.csv'
    df.to_csv(output_file, index=False)
```
